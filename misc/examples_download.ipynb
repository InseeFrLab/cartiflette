```python
from cartiflette.public import download_from_cartiflette
```

```python
ile_de_france = download_from_cartiflette(
      values = ["11"],
      crs = 4326,
      borders = "DEPARTEMENT",
      vectorfile_format="topojson",
      filter_by="REGION",
      source="EXPRESS-COG-CARTO-TERRITOIRE",
      year=2022,
      path_within_bucket = "test-download28")

ax = ile_de_france.boundary.plot(edgecolor = "k", linewidth  = 0.5)
ax.set_axis_off()
```

```python
ile_de_france = download_from_cartiflette(
      values = ["11","32","44"],
      crs = 4326,
      borders = "DEPARTEMENT",
      vectorfile_format="topojson",
      filter_by="REGION",
      source="EXPRESS-COG-CARTO-TERRITOIRE",
      year=2022,
      path_within_bucket = "test-download28")

ax = ile_de_france.plot(column = "INSEE_REG", edgecolor = "k", linewidth  = 0.5)
ax.set_axis_off()
```
